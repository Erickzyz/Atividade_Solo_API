```python
import requests
import pandas as pd

# Etapa 1: preparar o terreno (Lat e lon de ARUJA-SP)
lat =  -23.3965
lon = -46.3214
quantos_dias = 7

parametros = {
    'latitude': lat,
    'longitude': lon,
    'daily': ['temperature_2m_max', 'precipitation_sum'],
    'timezone': 'auto',
    'past_days': quantos_dias
}

# Etapa 2: Coleta de dados e verificação
url = "https://api.open-meteo.com/v1/forecast"
response = requests.get(url, params=parametros)

if response.status_code == 200:
  dados = response.json()

  dados_diarios = dados['daily']

  df_dados_clima = pd.DataFrame(dados_diarios)

  # Conversão de horario do EUA (Fahrenheit)
  df_dados_clima['temp_max_fahrenheit'] = (df_dados_clima['temperature_2m_max'] * 1.8) + 32

  print('Dadps coletados com sucesso!')
  print(df_dados_clima)

  # Etapa para salvar os Dados
  df_dados_clima.to_csv('relatorio_prof_luiz.csv', index = False)
  print("\nArquivo 'relatorio_prof_luiz.csv' salvo com sucesso")

else:
        print(f'Erro na conexão: {response.status_code}')
```
